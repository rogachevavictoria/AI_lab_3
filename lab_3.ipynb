```python
PokemonURL=os.path.join(data_dir,'Pokemon.csv')
PokemonURL
```

```python
# Read dataset
df = pd.read_csv(PokemonURL,engine='python', index_col=0)#to resolve UnicodeDecodeE
df.head()
```

```python
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
# Scatterplot arguments
sns.lmplot(x='Attack', y='Defense', data=df,
fit_reg=False, # No regression line
hue='Stage') # Color by evolution stage
```

```python
# Plot using Seaborn
sns.lmplot(x='Attack', y='Defense', data=df,
fit_reg=False,
hue='Stage')
# Tweak using Matplotlib
plt.ylim(0, None)
plt.xlim(0, None)
```

```python
# Boxplot
sns.boxplot(data=df)
```

```python
# Pre-format DataFrame
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
# New boxplot using stats_df
sns.boxplot(data=stats_df)
```

```python
sns.boxplot(x="Stage", y="Attack", data=df, palette="coolwarm")
```

```python
sns.boxplot(x="Stage", y="Attack",hue="Legendary", data=df, palette="coolwarm")
```

```python
# Set theme
sns.set_style('whitegrid')
# Violin plot
sns.violinplot(x='Type 1', y='Attack', data=stats_df)
```

```python
pkmn_type_colors = ['#78C850', # Grass
'#F08030', # Fire
'#6890F0', # Water
'#A8B820', # Bug
'#A8A878', # Normal
'#A040A0', # Poison
'#F8D030', # Electric
'#E0C068', # Ground
'#EE99AC', # Fairy
'#C03028', # Fighting
'#F85888', # Psychic
'#B8A038', # Rock
'#705898', # Ghost
'#98D8D8', # Ice
'#7038F8', # Dragon
]
# Violin plot with Pokemon color palette
sns.violinplot(x='Type 1', y='Attack', data=df,
palette=pkmn_type_colors) # Set color palette
```

```python
# Swarm plot with Pokemon color palette
sns.swarmplot(x='Type 1', y='Attack', data=df,palette=pkmn_type_colors)
```

```python
# Set figure size with matplotlib
plt.figure(figsize=(10,6))
# Create plot
sns.violinplot(x='Type 1',
y='Attack',
data=df,
inner=None, # Remove the bars inside the violins
palette=pkmn_type_colors)
sns.swarmplot(x='Type 1',
y='Attack',
data=df,
color='k', # Make points black
alpha=0.7) # and slightly transparent
# Set title with matplotlib
plt.title('Attack by Type')
```
